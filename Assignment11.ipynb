**Question 1: Write a basic Node.js script that prints "Hello, World!" to the console and show how to execute it.**

**Answer:**

1. Create a Node.js script

Create a file named hello.js and add the following code:

console.log("Hello, World!");

2. Execute the script

Open a terminal/command prompt, navigate to the folder containing hello.js, and run:

node hello.js

3. Output
Hello, World!


This confirms that Node.js is installed and your script executed successfully.

**Question 2: Create a simple CommonJS module that exports a function to add two numbers. Then, create a separate script that imports and uses this function.**

**Answer:**

1. Create a CommonJS module

Create a file named math.js and export a function that adds two numbers:

// math.js
function add(a, b) {
  return a + b;
}

module.exports = add;

2. Create a script that imports and uses the function

Create another file named app.js:

// app.js
const add = require("./math");

const result = add(5, 3);
console.log("Sum:", result);

**Question 3: Show the commands to initialize a new Node.js project and install the axios package as a dependency.**

**Answer:**

**Question 5: Using the fs/promises module, write a script to asynchronously read the content of a file named data.txt and log it to the console.**

**Answer:**

import { readFile } from "fs/promises";

async function readDataFile() {
  try {
    const data = await readFile("data.txt", "utf-8");
    console.log(data);
  } catch (error) {
    console.error("Error reading file:", error.message);
  }
}

readDataFile();

**Question 8: Create a basic HTTP server using the http module that listens on port 3000 and responds to all requests with "Server is running.**

**Answer:**

// server.js
const http = require("http");

const server = http.createServer((req, res) => {
  res.writeHead(200, { "Content-Type": "text/plain" });
  res.end("Server is running.");
});

server.listen(3000, () => {
  console.log("Server is listening on port 3000");
});

**Question 9: Enhance the previous HTTP server to handle two different routes: / should respond with "Homepage", and /about should respond with "About Us".**

**Answer:**

// server.js
const http = require("http");

const server = http.createServer((req, res) => {
  res.writeHead(200, { "Content-Type": "text/plain" });

  if (req.url === "/") {
    res.end("Homepage");
  } else if (req.url === "/about") {
    res.end("About Us");
  } else {
    res.writeHead(404, { "Content-Type": "text/plain" });
    res.end("Page Not Found");
  }
});

server.listen(3000, () => {
  console.log("Server is listening on port 3000");
});

**Question 10: Write a script that efficiently reads a large file (large.txt) and outputs its content to the console using streams.**

**Answer:**

// readStream.js
const fs = require("fs");

const readStream = fs.createReadStream("large.txt", {
  encoding: "utf8",
  highWaterMark: 64 * 1024 // 64 KB chunks
});

readStream.on("data", (chunk) => {
  console.log(chunk);
});

readStream.on("end", () => {
  console.log("Finished reading file.");
});

readStream.on("error", (err) => {
  console.error("Error reading file:", err.message);
});

**Question 1: Create a custom event emitter that fires a 'userLoggedIn' event with a username payload. Set up a listener to print a welcome message when the event occurs.**

**Answer:**

const EventEmitter = require("events");

// Create custom emitter
class UserEmitter extends EventEmitter {}

const userEmitter = new UserEmitter();

// Listener for userLoggedIn event
userEmitter.on("userLoggedIn", (username) => {
  console.log(`Welcome, ${username}!`);
});

// Emit the event with payload
userEmitter.emit("userLoggedIn", "Ankita");

**Question 2: Use the crypto module to create an SHA256 hash for the string "password123".**

**Answer:**

const crypto = require("crypto");

const hash = crypto
  .createHash("sha256")
  .update("password123")
  .digest("hex");

console.log(hash);

**Question 5: Write the minimal code for an Express.js server that listens on port 3000 and responds to GET requests at the root URL (/) with "Hello, Express!".**

**Answer:**

const express = require("express");
const app = express();

app.get("/", (req, res) => {
  res.send("Hello, Express!");
});

app.listen(3000, () => {
  console.log("Server running on port 3000");
});

**Question 6: Write a SQL statement to create a users table with an auto-incrementing primary key, an email, a password hash, and a creation timestamp.**

**Answer:**

CREATE TABLE users (
  id SERIAL PRIMARY KEY,
  email VARCHAR(255) NOT NULL UNIQUE,
  password_hash TEXT NOT NULL,
  created_at TIMESTAMP DEFAULT CURRENT_TIMESTAMP
);

**Question 10: Given a users table and a posts table (with a user_id foreign key), write a query to retrieve all posts and the email of the user who wrote them.**

**Answer:**

SELECT
    posts.*,
    users.email
FROM
    posts
JOIN
    users
ON
    posts.user_id = users.id;

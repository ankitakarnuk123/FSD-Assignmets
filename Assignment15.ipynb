**Question 1: What is Pug? Why might developers prefer Pug over writing raw HTML for server-rendered pages?**

**Answer:**

Pug (formerly known as Jade) is a template engine for Node.js used to generate HTML on the server side. It lets developers write HTML using a clean, indentation-based syntax instead of verbose HTML tags, and then compiles it into standard HTML before sending it to the browser.

Why developers might prefer Pug over raw HTML:

Less code, cleaner syntax
Pug removes closing tags and reduces boilerplate, making templates shorter and easier to read.

Dynamic content support
You can easily insert variables, conditions, and loops (like if, for, each) directly into the template.

Reusable components
Features like mixins, includes, and extends help avoid repetition and keep layouts modular.

Better maintainability
Large server-rendered apps become easier to manage due to structured layouts and reusable blocks.

Seamless integration with Express
Pug works smoothly with Express.js, making it a popular choice for server-side rendering in Node apps.

Built-in security features
By default, Pug escapes output, helping prevent XSS attacks unless explicitly disabled.

**Question 3: Explain what a RESTful API is. Then list four HTTP methods (GET, POST,PUT/PATCH, DELETE) and describe what each should do in a REST design for a resource**
â€œ/tasksâ€.

**Answer:**
A RESTful API (Representational State Transfer) is an API design style that allows clients (like web or mobile apps) to communicate with a server using standard HTTP protocols.
In REST, resources (such as users, tasks, or orders) are identified by URLs, and HTTP methods define the action to be performed on those resources. REST APIs are stateless, meaning each request contains all the information needed to process it.
HTTP Method	Endpoint	Purpose in REST	Description
GET	/tasks or /tasks/:id	Read	Retrieves tasks. /tasks fetches all tasks, /tasks/:id fetches a specific task.
POST	/tasks	Create	Creates a new task using the data sent in the request body.
PUT / PATCH	/tasks/:id	Update	Updates an existing task. PUT replaces the entire task, while PATCH updates only specific fields.
DELETE	/tasks/:id	Delete	Deletes the task with the given ID from the system.

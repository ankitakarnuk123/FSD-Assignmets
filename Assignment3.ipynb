**1.Question 1 : Create a layout with three boxes side by side using Flexbox. The first box should have twice the width of the other two. Provide the HTML and CSS code snippets.**
**Answer:**
<!DOCTYPE html>
<html lang="en">
<head>
  <meta charset="UTF-8">
  <meta name="viewport" content="width=device-width, initial-scale=1.0">
  <title>Flexbox Layout</title>
  <link rel="stylesheet" href="style.css">
</head>
<body>
  <div class="container">
    <div class="box box1">Box 1</div>
    <div class="box box2">Box 2</div>
    <div class="box box3">Box 3</div>
  </div>
</body>
</html>
<style>
.container {
  display: flex;          /* Enable Flexbox */
  gap: 10px;              /* Space between boxes (optional) */
}

.box {
  background-color: lightblue;
  padding: 20px;
  text-align: center;
  border: 2px solid #333;
}

/* First box should be twice the width */
.box1 {
  flex: 2;
}

.box2, .box3 {
  flex: 1;
}
</style>

  </div>

}

.box {
  background-color: lightcoral;
  padding: 20px;
  text-align: center;
  border: 2px solid #333;
}

/* First row spans across 2 columns */
.box1 {
  grid-column: 1 / span 2;
}

</style>

**Question 3: Write a media query that changes the background color of the body to light grey when the screen width is less than 600px.**
**Answer:**
<style>
body {
  background-color: white;
}

/* Media query for small screens */
@media (max-width: 600px) {
  body {
    background-color: lightgrey;
  }
}
</style>

**Question 6: Create a simple animation that moves a square box from left to right infinitely. Provide the HTML and CSS code snippets**
**Answer:**
<!DOCTYPE html>
<html lang="en">
<head>
  <meta charset="UTF-8">
  <meta name="viewport" content="width=device-width, initial-scale=1.0">
  <title>Box Animation</title>
  <link rel="stylesheet" href="style.css">
</head>
<body>
  <div class="box"></div>
</body>
</html>
<style>
.box {
  width: 50px;
  height: 50px;
  background-color: red;
  position: relative;
  animation: moveRight 3s linear infinite;
}

@keyframes moveRight {
  0% {
    left: 0;
  }
  50% {
    left: 300px;  /* Move to the right */
  }
  100% {
    left: 0;      /* Back to start */
  }
}

</style>

**Question 7: Explain the difference between position: absolute; and position: fixed; in CSS.**
**Answer:**
absolute: Positions the element relative to its nearest positioned ancestor and moves with the page when scrolled.

fixed: Positions the element relative to the browser window (viewport) and stays in place even when the page is scrolled.

**Question 8: Create a layered layout where a semi-transparent overlay covers an image using z-index. Provide the HTML and CSS code snippets.**
**Answer:**
<!DOCTYPE html>
<html lang="en">
<head>
  <meta charset="UTF-8">
  <meta name="viewport" content="width=device-width, initial-scale=1.0">
  <title>Image Overlay</title>
  <link rel="stylesheet" href="style.css">
</head>
<body>
  <div class="image-container">
    <img src="https://via.placeholder.com/600x400" alt="Sample Image">
    <div class="overlay"></div>
    <div class="text">Overlay Text</div>
  </div>
</body>
</html>
<style>
.image-container {
  position: relative;
  width: 600px;
  height: 400px;
}

.image-container img {
  width: 100%;
  height: 100%;
  display: block;
}

.overlay {
  position: absolute;
  top: 0;
  left: 0;
  width: 100%;
  height: 100%;
  background: rgba(0, 0, 0, 0.5); /* Semi-transparent black */
  z-index: 1;
}

.text {
  position: absolute;
  top: 50%;
  left: 50%;
  transform: translate(-50%, -50%);
  color: white;
  font-size: 24px;
  font-weight: bold;
  z-index: 2; /* Above overlay */
}
</style>
